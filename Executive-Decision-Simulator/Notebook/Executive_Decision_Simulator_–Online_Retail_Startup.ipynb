```python
# Simulate 12 months of SME data
months = pd.date_range(start='2026-01-01', periods=12, freq='M')

data = pd.DataFrame({
    'Month': months,
    'Revenue': np.random.randint(12000, 20000, size=12),
    'Cost': np.random.randint(7000, 13000, size=12),
    'Staff_Count': np.random.randint(5, 10, size=12),
    'Marketing_Spend': np.random.randint(2000, 5000, size=12),
    'Customer_Satisfaction': np.random.randint(70, 90, size=12),
    'Inventory_Level': np.random.randint(500, 1000, size=12),
    'Website_Traffic': np.random.randint(2000, 5000, size=12)
})
data['Profit'] = data['Revenue'] - data['Cost']
data.head()
```

```python
# Marketing Boost
marketing_scenario = data.copy()
marketing_scenario['Marketing_Spend'] *= 1.2
marketing_scenario['Revenue'] *= 1.1
marketing_scenario['Profit'] = marketing_scenario['Revenue'] - marketing_scenario['Cost']

# Staff Expansion
staff_scenario = data.copy()
staff_scenario['Staff_Count'] += 2
staff_scenario['Cost'] += 2*1500
staff_scenario['Customer_Satisfaction'] = np.minimum(staff_scenario['Customer_Satisfaction']*1.05, 100)
staff_scenario['Profit'] = staff_scenario['Revenue'] - staff_scenario['Cost']

# Cost Optimization
cost_scenario = data.copy()
cost_scenario['Cost'] *= 0.9
cost_scenario['Profit'] = cost_scenario['Revenue'] - cost_scenario['Cost']
```

```python
fig = px.line(data, x='Month', y='Profit', title='Profit Trends: Base vs Scenarios')
fig.add_scatter(x=marketing_scenario['Month'], y=marketing_scenario['Profit'], mode='lines', name='Marketing Boost')
fig.add_scatter(x=staff_scenario['Month'], y=staff_scenario['Profit'], mode='lines', name='Staff Expansion')
fig.add_scatter(x=cost_scenario['Month'], y=cost_scenario['Profit'], mode='lines', name='Cost Optimization')
fig.show()
```

```python
fig2 = px.line(data, x='Month', y='Customer_Satisfaction', title='Customer Satisfaction Trends')
fig2.add_scatter(x=staff_scenario['Month'], y=staff_scenario['Customer_Satisfaction'], mode='lines', name='Staff Expansion')
fig2.show()
```

```python
def scenario_score(df):
    return df['Profit']*0.6 + df['Customer_Satisfaction']*0.4

data['Score'] = scenario_score(data)
marketing_scenario['Score'] = scenario_score(marketing_scenario)
staff_scenario['Score'] = scenario_score(staff_scenario)
cost_scenario['Score'] = scenario_score(cost_scenario)

scores = {
    'Base': data['Score'].mean(),
    'Marketing Boost': marketing_scenario['Score'].mean(),
    'Staff Expansion': staff_scenario['Score'].mean(),
    'Cost Optimization': cost_scenario['Score'].mean()
}

best_strategy = max(scores, key=scores.get)
print("Best overall strategy:", best_strategy)
```

# Executive Decision Simulator – Online Retail Startup

This simulator models an online retail startup’s first year. Three strategic decisions were analyzed:

1. Marketing Boost: +20% marketing → +10% revenue
2. Staff Expansion: +2 staff → +5% customer satisfaction
3. Cost Optimization: -10% cost → higher profit

**Insights:**  
- Marketing improves revenue but slightly raises costs  
- Staff expansion boosts customer satisfaction  
- Cost optimization maximizes profit with minimal impact
